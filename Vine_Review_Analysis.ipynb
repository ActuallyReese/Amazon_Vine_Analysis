```python
import os
import csv
import pandas as pd
vine_csv = os.path.join("vine_table.csv")
```

```python
vine_table_df = pd.read_csv(vine_csv)
```

```python
vine_table_rows = vine_table_df[vine_table_df["total_votes"] >= 20]
vine_table_rows
```

```python
vine_table_helpful = vine_table_rows[vine_table_rows["helpful_votes"] / vine_table_rows["total_votes"] >= .5]
```

```python
vine_table_helpful
```

```python
vine_review_yes = vine_table_helpful[vine_table_helpful["vine"] == "Y"]
vine_review_no = vine_table_helpful[vine_table_helpful["vine"] == "N"]
```

```python
# Total number of reviews
yes_count = vine_review_yes["review_id"].count()
no_count = vine_review_no["review_id"].count()
print(f' Total vine_reviews: {yes_count}, Total non-vine reviews: {no_count}')
```

```python
yes_5 = vine_review_yes[vine_review_yes["star_rating"] == 5].star_rating.count()
no_5 = vine_review_no[vine_review_no["star_rating"] == 5].star_rating.count()
print(f' 5 star vine_reviews: {yes_5}, 5 star non-vine reviews: {no_5}')
```

```python
precision = 2
yes_percent = (yes_5 / yes_count) * 100
yes_percent = "{:.2f}".format(yes_percent)
no_percent = (no_5 / no_count) * 100
no_percent = "{:.2f}".format(no_percent)
str(f' Percent of 5 star vine_reviews: {yes_percent}, Percent of 5 star non-vine reviews: {no_percent}')
```
